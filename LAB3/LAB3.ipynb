```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
import random
```

```python
original = data.camera()
```

```python
plt.imshow(original, cmap = plt.cm.gray)
```

```python
image = data.camera()
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        image[i,j] += random.randint(-20,20)
```

```python
plt.imshow(image, cmap = plt.cm.gray)
```

```python
mm = np.median(image)
print(mm)
```

```python
binary = image > mm
```

```python
plt.imshow(binary, cmap = plt.cm.gray)
```

```python
from skimage import feature

edges1 = feature.canny(image,sigma=2)
```

```python
plt.imshow(edges1, cmap = plt.cm.gray)
```

```python
edges2 = feature.canny(image,sigma=3)
```

```python
plt.imshow(edges2, cmap = plt.cm.gray)
```

```python
edges3 = feature.canny(image,sigma=4)
```

```python
plt.imshow(edges3, cmap = plt.cm.gray)
```

```python
edges4 = feature.canny(image,sigma=11)
```

```python
plt.imshow(edges4, cmap = plt.cm.gray)
```

```python
edges5 = feature.canny(image,sigma=12)
```

```python
plt.imshow(edges5, cmap = plt.cm.gray)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=6, figsize=(15, 6))

ax[0].imshow(image, cmap='gray')
ax[0].set_title('noisy image', fontsize=15)

ax[1].imshow(edges1, cmap='gray')
ax[1].set_title(r'$\sigma=2$', fontsize=15)

ax[2].imshow(edges2, cmap='gray')
ax[2].set_title(r'$\sigma=3$', fontsize=15)

ax[3].imshow(edges3, cmap='gray')
ax[3].set_title(r'$\sigma=4$', fontsize=15)

ax[4].imshow(edges4, cmap='gray')
ax[4].set_title(r'$\sigma=11$', fontsize=15)

ax[5].imshow(edges5, cmap='gray')
ax[5].set_title(r'$\sigma=12$', fontsize=15)

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```
